# Manejo de módulos integrados

**NumPy:** Se usa principalmente para trabajar con datos numéricos de forma rápida, eficiente y estructurada. Un objeto tipo numpy es un arreglo base para hacer matemáticas y manipular datos en Python.

**Uso en ciencia de datos**


*   Manipulación de vectores y matrices
*   Álgebra lineal y cálculo matricial.
* Operaciones estadísticas y matemáticas.
* Procesamiento rápido de grandes volúmenes de datos numéricos.
* Base para librerías como TensorFlow, Scikit-learn y Pandas.

**Ventajas**
* Muy rápido y eficiente gracias a su implementación en C.
* Consume menos memoria que las listas tradicionales de Python.
* Facilita operaciones matemáticas complejas con pocas líneas de código.
* Ideal para algoritmos de
* Machine Learning y análisis numérico.
* Permite vectorización, evitando ciclos for lentos.

**Desventajas**
* Trabaja principalmente con datos numéricos homogéneos.
* Menos intuitivo para manipular datos tabulares.
* No posee herramientas avanzadas de limpieza o análisis de datos estructurados.
* Puede ser complejo para principiantes en álgebra matricial.

**Pandas:** Se usa para trabajar con datos estructurados (como tablas de Excel o bases de datos). Un objeto tipo pandas permite cargar, limpiar, transformar y analizar datos. La usamos para base de datos relacionales.

**Uso en ciencia de datos**

* Limpieza y transformación de datos.
* Lectura de archivos CSV, Excel o bases de datos.
* Filtrado, agrupación y análisis estadístico.
* Preparación de datos antes de entrenar modelos.
* Exploración y visualización preliminar de información.

**Ventajas**

* Muy fácil de usar para análisis de datos.
* Permite trabajar con columnas nombradas y distintos tipos de datos.
* Excelente para limpieza y preprocesamiento.
* Integra funciones estadísticas y de agrupación.
* Compatible con NumPy y Scikit-learn.

**Desventajas**

* Más lento que NumPy en cálculos numéricos masivos.
* Consume más memoria.
* No es la mejor opción para operaciones matemáticas muy complejas.
* Puede volverse ineficiente con datasets extremadamente grandes.

La diferencia entre **NumPy** y **Pandas** radica en su propósito y nivel de uso. NumPy es una librería enfocada en el cálculo numérico eficiente mediante arrays multidimensionales, ideal para operaciones matemáticas rápidas, álgebra lineal y procesamiento de datos puramente numéricos; mientras que Pandas es una librería de más alto nivel diseñada para la manipulación y análisis de datos estructurados, como tablas tipo Excel o bases de datos, permitiendo trabajar con columnas con nombres, datos de distintos tipos, limpieza, filtrado y agrupación; en la práctica, Pandas se apoya internamente en NumPy, por lo que ambos se complementan y se utilizan juntos en la mayoría de flujos de análisis de datos.

```python
import numpy as np


precios = np.array([12.50, 45.00, 8.90, 100.00, 15.50, 5.00])

# Método 1: np.sort() - Ordenar precios
precios_ordenados = np.sort(precios)
print("Precios de menor a mayor:", precios_ordenados)

# Método 2: np.mean() - Calcular el promedio
promedio = np.mean(precios)
print(f"Precio promedio: ${promedio:.2f}")

# Método 3: np.where() - Clasificar productos (Premium > $30)
clasificacion = np.where(precios > 30, 'Producto Premium', 'Producto Básico')
print("Clasificación de productos:", clasificacion)
```

```python
import pandas as pd

# Datos: Ventas semanales
data = {
    'Vendedor': ['Juan', 'Maria', 'Pedro', 'Ana'],
    'Ventas': [1500, 2200, 900, 3100],
    'Region': ['Norte', 'Sur', 'Norte', 'Sur']
}
df = pd.DataFrame(data)

# Método 1: groupby() - Sumar ventas por región
ventas_region = df.groupby('Region')['Ventas'].sum()
print("Total de ventas por región:")
print(ventas_region)

# Método 2: sort_values() - Ordenar por ventas
mejor_vendedor = df.sort_values(by='Ventas', ascending=False)
print("\nVendedores ordenados por ventas:")
print(mejor_vendedor)

# Método 3: describe() - Resumen estadístico
print("\nResumen estadístico de ventas:")
print(df['Ventas'].describe())
```

NumPy vs. Pandas en Ciencia de Datos

Para la implementación de algoritmos de Ciencia de Datos y Machine Learning, la elección entre estas dos librerías no es opcional, sino complementaria. A continuación, presento sus características clave bajo un enfoque de rendimiento y aplicabilidad:

NumPy: El motor de cálculo numérico

NumPy es el estándar para la computación intensiva debido a su capacidad de vectorización, lo que permite ejecutar operaciones sobre arreglos completos sin necesidad de usar bucles for lentos.

Ventajas:

Eficiencia en memoria: Almacena datos de forma contigua en la memoria, lo que optimiza el acceso y la velocidad de procesamiento (al estar compilado en C).

Base de datos científica: Es la capa sobre la que se construyen algoritmos complejos de álgebra lineal, transformadas de Fourier y generación de números aleatorios.

Desventajas:

Rigidez de tipo: Obliga a que todos los elementos de un arreglo sean del mismo tipo (ej. solo integers o solo floats). Esto lo hace ineficiente para datos del mundo real que suelen ser "sucios" y mezclan etiquetas de texto con valores numéricos.

Falta de contexto: Los arreglos de NumPy no tienen nombres de columnas ni índices significativos, lo que obliga al programador a recordar qué significa cada posición o columna, aumentando el riesgo de errores lógicos.

Pandas: La herramienta de estructuración y limpieza

Pandas extiende la potencia de NumPy para trabajar con DataFrames, estructuras tabulares (similares a Excel o SQL) que permiten etiquetar datos.

Ventajas:

Gestión de datos heterogéneos: Permite manipular en una misma tabla números, fechas, categorías y texto, facilitando el trabajo con datasets del mundo real.

Potencia en limpieza: Incluye métodos nativos para gestionar datos faltantes (NaN), realizar joins (uniones) entre tablas y agrupar información mediante groupby, algo que sería extremadamente complejo y tedioso de programar desde cero en NumPy.

Desventajas:

Costo computacional: Debido a la flexibilidad y la sobrecarga que implican los índices y metadatos del DataFrame, Pandas es significativamente más lento y consume más memoria RAM que NumPy.

Limitaciones en Big Data: Al cargar todo el conjunto de datos en la memoria activa del equipo, no es apto para procesar archivos que excedan la capacidad de la memoria RAM disponible, obligando a buscar alternativas como el procesamiento distribuido en Spark.
